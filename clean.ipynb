```python
# === Shopping List Manager (clean) ===
# RULES: Use only Week 1â€“3 features: variables, input(), print(), if/elif/else, while.
#        No lists/dicts/functions/files. We represent the shopping list as 10 separate slots.
# LIMIT: Up to 10 items. Empty slot = "" (nothing there).

# --- item slots ---
slot1 = ""
slot2 = ""
slot3 = ""
slot4 = ""
slot5 = ""
slot6 = ""
slot7 = ""
slot8 = ""
slot9 = ""
slot10 = ""

running = True
while running:
    print("=== Shopping List Manager ===")
    print("1. Add item")
    print("2. View items")
    print("3. Remove item")
    print("4. Count items")
    print("5. Exit")
    choice = input("Choose option: ").strip()

    if choice == "1":
        # Add new item to the first empty slot
        item = input("Enter item: ").strip()
        added = False

        if slot1 == "":
            slot1 = item; added = True
        elif slot2 == "":
            slot2 = item; added = True
        elif slot3 == "":
            slot3 = item; added = True
        elif slot4 == "":
            slot4 = item; added = True
        elif slot5 == "":
            slot5 = item; added = True
        elif slot6 == "":
            slot6 = item; added = True
        elif slot7 == "":
            slot7 = item; added = True
        elif slot8 == "":
            slot8 = item; added = True
        elif slot9 == "":
            slot9 = item; added = True
        elif slot10 == "":
            slot10 = item; added = True

        if added:
            print("Item added!")
        else:
            print("List is full (10 items). Remove something first.")

    elif choice == "2":
        # Show items with numbers (1..N) in the order of slots.
        print("Current items:")
        shown_any = False
        number = 0
        if slot1 != "":
            number = number + 1; print(str(number) + ". " + slot1); shown_any = True
        if slot2 != "":
            number = number + 1; print(str(number) + ". " + slot2); shown_any = True
        if slot3 != "":
            number = number + 1; print(str(number) + ". " + slot3); shown_any = True
        if slot4 != "":
            number = number + 1; print(str(number) + ". " + slot4); shown_any = True
        if slot5 != "":
            number = number + 1; print(str(number) + ". " + slot5); shown_any = True
        if slot6 != "":
            number = number + 1; print(str(number) + ". " + slot6); shown_any = True
        if slot7 != "":
            number = number + 1; print(str(number) + ". " + slot7); shown_any = True
        if slot8 != "":
            number = number + 1; print(str(number) + ". " + slot8); shown_any = True
        if slot9 != "":
            number = number + 1; print(str(number) + ". " + slot9); shown_any = True
        if slot10 != "":
            number = number + 1; print(str(number) + ". " + slot10); shown_any = True
        if shown_any == False:
            print("(no items yet)")

    elif choice == "3":
        # Remove by the displayed number. We'll count through the non-empty slots
        # until we reach the target number, then blank that slot.
        print("Tip: Choose 'View items' first to see numbers.")
        num_text = input("Number to remove: ").strip()

        is_digit = (num_text != "" and num_text[0] >= "0" and num_text[0] <= "9")
        if is_digit:
            target = int(num_text)
            count = 0

            if slot1 != "":
                count = count + 1
                if count == target:
                    slot1 = ""
            if slot2 != "":
                count = count + 1
                if count == target:
                    slot2 = ""
            if slot3 != "":
                count = count + 1
                if count == target:
                    slot3 = ""
            if slot4 != "":
                count = count + 1
                if count == target:
                    slot4 = ""
            if slot5 != "":
                count = count + 1
                if count == target:
                    slot5 = ""
            if slot6 != "":
                count = count + 1
                if count == target:
                    slot6 = ""
            if slot7 != "":
                count = count + 1
                if count == target:
                    slot7 = ""
            if slot8 != "":
                count = count + 1
                if count == target:
                    slot8 = ""
            if slot9 != "":
                count = count + 1
                if count == target:
                    slot9 = ""
            if slot10 != "":
                count = count + 1
                if count == target:
                    slot10 = ""

            print("Removed (if that number existed).")
        else:
            print("Please type a number like 1, 2, or 3.")

    elif choice == "4":
        # Count how many slots are non-empty
        total = 0
        if slot1 != "": total = total + 1
        if slot2 != "": total = total + 1
        if slot3 != "": total = total + 1
        if slot4 != "": total = total + 1
        if slot5 != "": total = total + 1
        if slot6 != "": total = total + 1
        if slot7 != "": total = total + 1
        if slot8 != "": total = total + 1
        if slot9 != "": total = total + 1
        if slot10 != "": total = total + 1
        print("Total items:", total)

    elif choice == "5":
        print("Goodbye!")
        running = False
    else:
        print("Please choose a number 1-5.")
```
